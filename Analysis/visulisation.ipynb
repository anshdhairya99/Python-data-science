```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly. express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_length_mm']
```

```python
penguins['bill_length_mm'].plot()
```

```python
penguins['bill_depth_mm']
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='',
                               marker='x',
                               markersize=5,
                               alpha=0.7,
                               figsize=(5,5),
                               markerfacecolor='yellow',
                               markeredgecolor='black',)
                               
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,2), color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,2), color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,4),
                                     color='blue',
                                       alpha=0.5)
```

```python
# it is used for machine learning
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm',s = 100, alpha=0.5)
```

```python
penguins.plot(kind='scatter', x='bill_length_mm', y='flipper_length_mm',s = 100, alpha=0.5)
```

- multiple graph

- Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 4), constrained_layout= True )
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 4), constrained_layout= True )
penguins.plot(kind='scatter', x='bill_length_mm', y='flipper_length_mm',s = 100, alpha=0.5, ax=axs[0], title='penguin bill data')
penguins.plot(kind='scatter', x='bill_length_mm', y='flipper_length_mm',s = 100, alpha=0.5,ax=axs[1], title='penguin bill data')
penguins.plot(kind='scatter', x='bill_length_mm', y='flipper_length_mm',s = 100, alpha=0.5,ax=axs[1], title='penguin flipper')


plt.show()
```
